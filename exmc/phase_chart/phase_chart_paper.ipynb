### spc3 vs chi1

```julia
# binの調整
# n:サンプルサイズ
# k:binの個数
# h:bin幅

k(n) = 1 + log2(n)
h(sigma,n) = 3.5 * sigma / n^(1/3)

function bin_width(int_data)
    sigma_int = std(int_data[:,end])
    n_sample = length(int_data[:,end])

    #bin幅
    h_width = h(sigma_int, n_sample)
    return h_width
end
```

```julia
plt = plot(
    chi1[:,end],
    st = :histogram,
    label="mag",
    alpha = 0.5,
    #bin = b_range1,
)
plt = plot!(
    spc3[:,end],
    st = :histogram,
    label="spc",
    alpha = 0.5,
    #bin = b_range1,
)
plt = plot!(
    [B40_true],
    seriestype = "vline",
    linestyle = :dash,
    linewidth = 2,
    color = :black,
    label = "",
)

```

```julia
h_bin1 = bin_width(int1[:,end])
bin_min1 = minimum(chi1[:,end])
bin_max1 = maximum(chi1[:,end])

b_range1 = collect(bin_min1:h_bin1:bin_max1);
```

```julia
plt_1 = plot(
    chi1[:,end],
    st = :histogram,
    label="mag",
    alpha = 0.5,
    bin = b_range1,
)
plt_1 = plot!(
    spc3[:,end],
    st = :histogram,
    label="spc",
    alpha = 0.5,
    bin = b_range1,
)
plt_1 = plot!(
    [B40_true],
    seriestype = "vline",
    linestyle = :dash,
    linewidth = 2,
    color = :black,
    label = "",
)


plt_2 = plot(
    int1[:,end],
    st = :histogram,
    label="int",
    alpha = 0.5,
    bin = b_range1,
    color = palette(:tab10)[3],
)
plt_2 = plot!(
    [B40_true],
    seriestype = "vline",
    linestyle = :dash,
    linewidth = 2,
    color = :black,
    label = "",
)


plts = plot(
    plt_1,
    plt_2,
    framestyle = :box,
    xlabel = L"B40",
    ylabel = "frequency",
    layout = (2, 1),
    grid = false,
    xtickfont = font(13),
    ytickfont = font(13),
    guidefont = font(13),
    legendfont = font(13),
    size=(600,800),
    ylims = (0,180),
    #xlims = (0.1135,0.1199),
    margin=Plots.Measures.Length(:mm, 3.0),
)
```

### free energy

```julia
free_spc_3 = -703.6346177171628
free_spc_4 = -304.5950188268978
free_spc_5 = 96.29764776950208
```

```julia
free_int_1 = -1011.0814952463504
free_int_2 = -210.5721212144229
free_int_3 = 591.5807528325613
```

```julia
free_chi_1 = -292.45684243563323
free_chi_2 = 98.53070126539782
free_chi_3 = 500.9385115097629
```

```julia
free_spc = [
    free_spc_3,
    free_spc_4,
    free_spc_5,
]
```

```julia
free_chi = [
    free_chi_1,
    free_chi_2,
    free_chi_3,
]
```

```julia
free_int = [
    free_int_1,
    free_int_2,
    free_int_3,
]
```

```julia
free_sum = free_chi + free_spc
```

```julia
plot(
    free_sum,
    color = "black",
    markershape = :circle,
    label = L"F_{\mathrm{mag}}+F_{\mathrm{spc}}",
)

plot!(
    free_int,
    color = "orange",
    markershape = :diamond,
    label = L"F_{\mathrm{int}}",
)
```

```julia
dif = free_int - free_sum
```

```julia
prob_int = 1 ./ (1 .+ exp.(dif))
```

```julia
prob_sum = 1 .- prob_int
```

```julia
plot(
    bar(
        ["INT","MAG+SPC"],
        #[1,2],
        [prob_int[1],prob_sum[1]],
        ylabel = L"p(M|D)",
        label = "Likelyhood",
        color = :gray,
        alpha = 0.5,
        legend=:topleft,
    )
)

plot!(
    twinx(),
    ["INT","MAG+SPC"],
    [free_int[1], free_sum[1]],
    markershape=:circle,
    color = :red,
    ylabel="Free Energy",
    label="Free Energy",
    legend=:topright, 
    ylims = (-1015,-990),
)

plot!(
    grid=false,
    framestyle = :box,
)
```
