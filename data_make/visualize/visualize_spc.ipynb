```julia
#package
using Distributions
using Random
using LinearAlgebra
using Plots
using StatsBase
using LaTeXStrings
using DataFrames
using CSV
using Base.Threads
using HDF5
gr()
```

```julia
function read_file(file_path)
    file = h5open(file_path, "r")
    SpcHeat_Temp = read(file, "spc")
    return SpcHeat_Temp
end
```

```julia
function visualize(data)
    # 温度設定
    Temp_table_spc = collect(0.2:0.2:70)
    plot(
        Temp_table_spc, data,
        color="black",
        xlabel=L"$T$ (K)",
        ylabel=L"$C$ (J(mol $\cdot$ K))",
        label="Observed data",
        st=:scatter,
        markersize=0.5,
        framestyle = :box,
        title = L"\sigma=0",
    )
end
```

```julia
read_file_path = "/Users/test/home/lab_research_1/data_make/data/spc/spc.h5"
SpcHeat_Temp = read_file(read_file_path)
visualize(SpcHeat_Temp)
```

```julia
# pathの設定
save_main_path = "/Users/test/home/lab_research_1/data_make/visualize/spc/"

save_png = ".png"
save_pdf = ".pdf"

individual_path = "spc"

save_path_png = save_main_path * individual_path * save_png
save_path_pdf = save_main_path * individual_path * save_pdf

# 保存
savefig(save_path_png)
savefig(save_path_pdf)
```

```julia
function save_path_return(index)
    # pathの設定
    save_main_path = "/Users/test/home/lab_research_1/data_make/visualize/spc/spc_"

    # pdf, pngの拡張子のパス
    save_png = ".png"
    save_pdf = ".pdf"

    individual_path = "spc"

    save_path_png = save_main_path * string(index) * save_png
    save_path_pdf = save_main_path * string(index) * save_pdf

    return save_path_png, save_path_pdf
end
```
